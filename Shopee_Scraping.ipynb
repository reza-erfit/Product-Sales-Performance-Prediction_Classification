```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import requests
import time
```

```python
#Link Digimap
main_link = 'https://shopee.co.id/search?keyword=lemari'
path = 'C:/Users/Lenovo/OneDrive/Dokumen/Web Driver/chromedriver.exe'

# Customize chrome display
chrome_options = Options()
chrome_options.add_argument('--no-sandbox')
#chrome_options.add_argument('--headless')
chrome_options.add_argument('disable-notifications')
chrome_options.add_argument('--disable-infobars')

driver = webdriver.Chrome(executable_path = path, options=chrome_options)
driver.get(main_link)
```

```python
html = driver.execute_script("return document.getElementsByTagName('html')[0].innerHTML")
soup = BeautifulSoup(html,"html.parser")
```

```python
itemid,shopid,name ,stock, historical_sold, liked_count, price, price_min, price_max,price_min_before_discount, price_max_before_discount, price_before_discount, raw_discount, rating_star, show_free_shipping, is_preferred_plus_seller, shop_location, can_use_cod, is_on_flash_sale, rating_total,rating_count_1, rating_count_2, rating_count_3,rating_count_4,rating_count_5 =  [], [], [], [], [], [],[],[], [],[], [], [], [], [], [], [], [], [], [], [], [], [], [], [], []
shopee_url = 'https://shopee.co.id'
keyword = 'lemari'
newest = 0

while newest < 3001 :
    header = {
        'User-Agent' : 'Chrome', 
        'Referer' : '{}search?keyword={}'.format(shopee_url, keyword)}

    url = 'https://shopee.co.id/api/v4/search/search_items?by=relevancy&keyword={}&limit=100&newest={}&order=desc&page_type=search&scenario=PAGE_GLOBAL_SEARCH&version=2'.format(keyword, newest)


    # API request
    r = requests.get(url, headers = header).json()

    #Scraping

    for item in r['items'] :
        itemid.append(item['item_basic']["itemid"])
        shopid.append(item['item_basic']["shopid"])
        name.append(item['item_basic']["name"])
        stock.append(item['item_basic']["stock"])
        historical_sold.append(item['item_basic']["historical_sold"])
        liked_count.append(item['item_basic']["liked_count"])
        price.append(item['item_basic']["price"])
        price_min.append(item['item_basic']["price_min"])
        price_max.append(item['item_basic']["price_max"])
        price_min_before_discount.append(item['item_basic']["price_min_before_discount"])
        price_max_before_discount.append(item['item_basic']["price_max_before_discount"])
        price_before_discount.append(item['item_basic']["price_before_discount"])
        raw_discount.append(item['item_basic']["raw_discount"])
        rating_star.append(item['item_basic']["item_rating"]["rating_star"])
        show_free_shipping.append(item['item_basic']["show_free_shipping"])
        is_preferred_plus_seller.append(item['item_basic']["is_preferred_plus_seller"])
        shop_location.append(item['item_basic']["shop_location"])
        can_use_cod.append(item['item_basic']["can_use_cod"])
        is_on_flash_sale.append(item['item_basic']["is_on_flash_sale"])
        rating_total.append(item['item_basic']["item_rating"]["rating_count"][0])
        rating_count_1.append(item['item_basic']["item_rating"]["rating_count"][1])
        rating_count_2.append(item['item_basic']["item_rating"]["rating_count"][2])
        rating_count_3.append(item['item_basic']["item_rating"]["rating_count"][3])
        rating_count_4.append(item['item_basic']["item_rating"]["rating_count"][4])
        rating_count_5.append(item['item_basic']["item_rating"]["rating_count"][5])
    newest +=30
```

```python
#SAVE DATA
listCols = ["itemid","shopid","name","stock", "historical_sold", "liked_count","price","price_min", "price_max", "price_min_before_discount", "price_max_before_discount", "price_before_discount", "raw_discount", "rating_star", "show_free_shipping", "is_preferred_plus_seller", "shop_location", "can_use_cod", "is_on_flash_sale",  "rating_total","rating_count_1", "rating_count_2","rating_count_3","rating_count_4","rating_count_5"]

dict_data = dict(zip(
            listCols,
                (itemid,
                shopid,
                name, 
                stock,  
                historical_sold, 
                liked_count,  
                price, price_min, 
                price_max, 
                price_min_before_discount, 
                price_max_before_discount, 
                price_before_discount, 
                raw_discount, rating_star, 
                show_free_shipping, 
                is_preferred_plus_seller, 
                shop_location,
                can_use_cod, 
                is_on_flash_sale,
                rating_total,
                rating_count_1, 
                rating_count_2,
                rating_count_3,
                rating_count_4,
                rating_count_5)))
```

```python
import json
with open('lemari.json','w') as fp:
    json.dump(dict_data, fp)
```
